```python
import xml.etree.cElementTree as et
 
parsedXML = et.parse( "Restaurants_Train_Final.xml" )
```

```python
def gett(node):
    """ return node text or None """
    return node if node is not None else None

i=0
import pandas as pd
df=pd.DataFrame({'review':[],"aspect":[],"polarity":[]}) 
reviews=parsedXML.getroot()
for review in reviews:
    for sents in review:
        for sentid in sents:
            fentry=opinion=pole="NONE"
            for txt in sentid:
                fetry=txt.text
                i+=1
                for op in txt:
                    opinion=op.attrib.get('category')
                    pole=op.attrib.get('polarity')
                df.loc[i]=[gett(fetry),gett
                           (opinion),gett(pole)]              
```
